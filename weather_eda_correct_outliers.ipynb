```python
import numpy as np
import pandas as pd
from scipy import stats
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('weather_filled.csv', delimiter=',', parse_dates=True, index_col='Date Time')
df.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
for col in df.columns.tolist():
    x = np.sort(np.asarray(df[col].index))
    y = np.asarray(df[col].values)
    fig = plt.figure(figsize=(20, 2))
    fig.suptitle(col)
    ax = fig.add_subplot(111)
    ax.plot(x, y)
```

```python
saale2015 = pd.read_csv('mpi_saale_2015b.csv', delimiter=',', parse_dates=True, index_col='Date Time')
saale2015.shape
```

```python
df.loc['2015-07-13 09:10:00':
       '2015-07-13 12:00:00', 'wv (m/s)'] = saale2015.loc['2015-07-13 09:10:00':
                                                          '2015-07-13 12:00:04', 'wv (m/s)'].values
```

```python
df.loc['2015-07-13 09:00:00':
       '2015-07-13 12:10:00', 'max. wv (m/s)'] = saale2015.loc['2015-07-13 09:00:04':
                                                               '2015-07-13 12:10:04', 'wv (m/s)'].values
```

```python
x = np.sort(np.asarray(df['wv (m/s)'].index))
y = np.asarray(df['wv (m/s)'].values)
fig = plt.figure(figsize=(20, 2))
fig.suptitle('wv (m/s)')
ax = fig.add_subplot(111)
ax.plot(x, y)
```

```python
x = np.sort(np.asarray(df['max. wv (m/s)'].index))
y = np.asarray(df['max. wv (m/s)'].values)
fig = plt.figure(figsize=(20, 2))
fig.suptitle('2015 wv (m/s)')
ax = fig.add_subplot(111)
ax.plot(x, y)
```
