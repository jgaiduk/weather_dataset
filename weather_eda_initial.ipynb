```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing Jena Weather dataset from Max-Plank Institute weather station. This dataset includes entries from Beutenberg WS. Separate zip archives can be downloded from https://www.bgc-jena.mpg.de/wetter/weather_data.html

```python
data = pd.read_csv('weather.csv', delimiter=',')
data.describe()
```

```python
data.isnull().sum()
```

No null values. Check for duplicate entries. 

```python
data_duplicated = data[data.duplicated(['Date Time'], keep=False)] 
# keep=False indicates that we want to display all duplicate entries
data_duplicated.sort_values(by='Date Time')
```

```python
# All duplicate entries are grouped by datetime value
value_counts = data_duplicated['Date Time'].value_counts()
value_counts[:5]
```

```python
# check whether all value groups have only 2 entries each
value_counts.nunique() == 1
```

```python
# we can safely drop duplicates 
data_no_duplicates = data.drop_duplicates()
```

```python
# sanity check 
data.shape[0] - data_duplicated.shape[0] / 2 == data_no_duplicates.shape[0]
```

```python
data.to_csv('weather_nodup.csv', sep=',', index=False)
```
